```python
import pandas as pd
# 1 读取数据
train_id = []
train_text = []
with open('./Train_DataSet.csv',encoding='utf8') as f:
    for idx,i in enumerate(f):
        if idx == 0:
            pass
        else:
            ik = str(i).split(',')[0]
            i = str(i).split(',')[1:]
            train_id.append(ik)
            train_text.append(','.join(i).replace('\\n',' ').replace('\\r\\n','').replace(' ','').replace('.',''))
train = pd.DataFrame()
train['id'] = train_id
train['text'] = train_text
train_label = pd.read_csv('./Train_DataSet_Label.csv',sep=',')
train = pd.merge(train,train_label,on=['id'],copy=False)

print(train.info())
train.head()
train.to_csv("../input/Train_DataSet.csv",index = False)
```

```python
test_id = []
test_text = []
with open('./Test_DataSet.csv',encoding='utf8') as f:
    for idx,i in enumerate(f):
        if idx == 0:
            pass
        else:
            ik = str(i).split(',')[0]
            i = str(i).split(',')[1:]
            test_id.append(ik)
            test_text.append(','.join(i).replace('\\n',' ').replace('\\r\\n','').replace(' ','').replace('.',''))
test = pd.DataFrame()
test['id'] = test_id
test['text'] = test_text
test.to_csv("../input/Test_DataSet.csv",index = False)
```

```python
import pandas as pd
train_data = []
id_set = set()
with open("./Train_DataSet.csv",encoding="utf-8") as f:
    f.readline()
    for line in f:
        if len(line.split(",")) != 3:
            print(line)
        nid = line.split(",")[0]
        title = line.split(",")[1].replace('\\n',' ').replace('\\r\\n','').replace(' ','').replace('.','')
        content = line.split(",")[2].replace('\\n',' ').replace('\\r\\n','').replace(' ','').replace('.','').replace("???",'')
        id_set.add(nid)
        train_data.append([nid,title,content])
train_data
```

```python
train_x = pd.DataFrame(train_data,columns=["id","title","content"])
train_y = pd.read_csv("./Train_DataSet_Label.csv")
train_x.info()
```

```python
train_data = pd.merge(train_x,train_y,on=['id'],copy=False)
train_data.to_csv("./Train_DataSet_title_content.csv",index = False)
```
